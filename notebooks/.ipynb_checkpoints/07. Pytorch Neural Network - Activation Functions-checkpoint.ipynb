# ACTIVATION FUNCTIONS

### Replicating non-linear activation functions
In Neural Network, we harness the power of discovering patterns in data by giving the models some tools rather then just telling the model what to learn.

We call these tools **Activation Functions**, a combination of linear and non-linear functions.

Let's see some common activation functions we have already worked on and some other too.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

First, we create a linear data to help visualize the activation functions for us.

```python
# Create a tensor
A = torch.arange(-10,10, 1, dtype=torch.float32)
A.dtype
```

```python
plt.plot(A)
```

## ReLU (Rectified Linear Unit) Activation Function

Rectified Linear Unit (ReLU) is a popular activation functions used in neural networks, specially in deep learning. The ReLU function is a piecewise linear function that outputs the input directly if it is positive; otherwise, it outputs zero. In simpler terms, ReLU allows positive values to pass through unchanged while setting all negative values to zero. It is more simple than other actiation functions and avoids the complexity of vanishing gradient.

$ReLU(x) = f(x) = max(0, x)$

$\text{ReLU}(x) = \begin{cases} 
x & \text{if } x \geq 0 \\ 
0 & \text{if } x < 0 
\end{cases}$

```python
# Pytorch's in-built ReLU
plt.plot(torch.relu(A))
```

```python
""" Replicating the ReLU activation by ourself"""
def relu(x: torch.Tensor) -> torch.Tensor:
    return torch.maximum(torch.tensor(0), x) # Input must be tensors

relu(A)
```

```python
# Our ReLU activation
plt.plot(relu(A))
```

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
plt.plot(torch.sigmoid(A))
```

```python
plt.plot(sigmoid(A))
```
