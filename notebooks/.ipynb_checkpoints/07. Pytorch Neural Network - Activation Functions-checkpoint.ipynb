In Neural Network, we harness the power of discovering patterns in data by giving the models some tools rather then just telling the model what to learn.

We call these tools **Activation Functions**, a combination of linear and non-linear functions.

Let's see some common activation functions we have already worked on and some other too.

1. **Rectified Linear Unit - ReLU `torch.nn.ReLU()`**
2. **Sigmoid `torch.nn.Sigmoid()`**
3. **Hyperbolic Tangent - Tanh `torch.nn.Tanh()`**

#### Create a Data for Activation Function Visualization
First, we create a linear data to help visualize the activation functions for us.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a tensor
A = torch.arange(-20,20, 0.1, dtype=torch.float32)
print(f"Data Type: {A.dtype}")
print(f"Length: {len(A)}")
```

```python
# Let's plot and see if the data is linear or not.
plt.plot(A)
```

## Rectified Linear Unit (ReLU) Activation Function

Rectified Linear Unit (ReLU) is a popular activation functions used in neural networks, specially in deep learning. The ReLU function is a piecewise linear function that outputs the input directly if it is positive; otherwise, it outputs zero. In simpler terms, ReLU allows positive values to pass through unchanged while setting all negative values to zero. It is more simple than other activation functions and avoids the complexity of vanishing gradient.

$ReLU(x) = f(x) = max(0, x)$

$\textit{ReLU}(x) = \begin{cases} 
x & \text{if } x \geq 0 \\ 
0 & \text{if } x < 0 
\end{cases}$

### ReLU using PyTorch

```python
# Let's plot Pytorch's in-built ReLU
plt.title("ReLU Activation with PyTorch")
plt.plot(torch.relu(A))
```

### Replicating ReLU Activation

```python
""" Replicating the ReLU activation by ourself"""
def relu(x: torch.Tensor) -> torch.Tensor:  # Return must be tensors
    return torch.maximum(torch.tensor(0), x) # Input must be tensors
```

```python
# Plot our ReLU activation
plt.title("Replicated ReLU Activation")
plt.plot(relu(A))
```

## Sigmoid Activation Function

A sigmoid activation function in a neural network is a mathematical function that transforms a neuron's input value into an output between 0 and 1. The ouput shape looks like S-shape, which is called the sigmoidal curve, making it particularly useful for binary classification problems where the output can be interpreted as a probability. It is calculated by an exponential function.

$ Sigmoid(x) = \sigma(x) = \frac{1}{1 + e^{-x}}$

$
\textit{Sigmoid(x)} = 
\begin{cases} 
1 & \text{if } x \to \infty \\
0.5 & \text{if } x = 0 \\
0 & \text{if } x \to -\infty \\
\frac{1}{1 + e^{-x}} & \text{otherwise}
\end{cases}
$

### Sigmoid using PyTorch

```python
# Let's plot Pytorch's in-built Sigmoid
plt.title("Sigmoid Activation with PyTorch")
plt.plot(torch.sigmoid(A))
```

### Replicating Sigmoid Activation

```python
""" Replicating the Sigmoid activation by ourself"""
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
# Plot our Sigmoid activation
plt.title("Replicating Sigmoid Activation")
plt.plot(sigmoid(A))
```

## Hyperbolic Tangent (Tanh) Activation Function

The Tanh activation function, or hyperbolic tangent, is a non-linear function that maps input values to the range of -1 to 1, unlike the sigmoid function, which maps to 0 to 1. It enables neural networks to learn complex patterns. The output values are centered around zero, which can facilitate learning in subsequent layers. Tanh can lead to faster convergence compared to sigmoid, especially in multi-layer neural networks. Also, it can sometimes achieve better performance than sigmoid for multi-layer neural networks. But Tanh, like sigmoid, can suffer from the vanishing gradient problem, particularly in deep networks.

$ tanh(x) = \frac{(e^x - e^{-x})}{(e^x + e^{-x})} $

$ \tanh(x) =
\begin{cases} 
  1 & \text{if } x \to \infty, \\
  \frac{e^x - e^{-x}}{e^x + e^{-x}} & \text{if } x \in \mathbb{R}, \\
  -1 & \text{if } x \to -\infty.
\end{cases} $

### Tanh using PyTorch

```python
# Let's plot Pytorch's in-built Tanh
plt.title("Tanh Activation with PyTorch")
plt.plot(torch.tanh(A))
```

### Replicating Tanh Activation

```python
""" Replicating the Tanh activation by ourself"""
def Tanh(x):
    return ((torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x)))
```

```python
# Plot our Tanh activation
plt.title("Replicating Tanh Activation")
plt.plot(Tanh(A))
```

## Softmax Activation Function

The softmax function, also known as *softargmax* or *normalized exponential function*, converts a vector of raw prediction scores (logits) into probabilities. More specifically, it converts a vector of K real numbers into a probability distribution of K possible outcomes. This process ensures the output values are in the range (0,1) and sum up to 1, making them interpretable as probabilities. 

Softmax is ideal for problems involving more than two classes, where the goal is to predict a single class out of many. ince Softmax converts logits into probabilities, the output is easily interpretable. You can not only determine the most likely class but also gauge the confidence of the model in that prediction. This holistic approach ensures that each prediction takes into account all classes rather than just focusing on one or two.

$ \textit{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^K e^{x_j}}, \quad \text{for } i = 1, 2, \dots, K. $

$ \textit{Softmax}(x_i) =
\begin{cases} 
  0 & \text{if } x_i \text{ is not the maximum element,} \\
  \frac{e^{x_i}}{\sum_{j=1}^K e^{x_j}} & \text{if } x_i \in \mathbb{R}, \\
  1 & \text{if } x_i \text{ is the maximum element.}
\end{cases} $

### Softmax using PyTorch

```python
# Let's plot Pytorch's in-built Softmax
plt.title("Softmax Activation with PyTorch")
plt.plot(torch.softmax(A, dim=0))
```

### Replicating Softmax Activation

```python
def Softmax(x):
    exponents = torch.exp(x)  # Compute all exponentials at once
    sum_of_exponents = torch.sum(exponents)  # Sum of all exponentials
    probabilities = exponents / sum_of_exponents  # Element-wise division
    return probabilities
```

```python
# Plot our Softmax activation
plt.title("Replicating Softmax Activation")
plt.plot(Softmax(A))
```

## Scaled Exponential Linear Unit (SELU) Activation Function

The Scaled Exponential Linear Unit (SELU) is a self-normalizing activation function that aims to stabilize neural network training by promoting a zero mean and unit variance in the outputs of each layer, potentially reducing the need for batch normalization. SELU's primary advantage is its self-normalizing property. This means that the outputs of each layer tend to converge towards a zero mean and unit variance, making the training process more stable and potentially faster. 

SELU's self-normalizing behavior can help stabilize the training process, particularly in deep networks, where gradients can become problematic. Because SELU's outputs are already standardized, using batch normalization might be less necessary. The self-normalizing property can also help to improve the model's generalization ability. 

$ \textit{SELU}(x) = \lambda \cdot ( max (0, x))+ min(0, \alpha \cdot (e^x - 1))$

$ \text{SELU}(x) =
\begin{cases} 
\lambda \cdot x & \text{if } x > 0, \\
\lambda \cdot \alpha \cdot (e^x - 1) & \text{if } x \leq 0.
\end{cases}$

$ \quad \text{where, } \textit{Scaling Factor (} \lambda \textit{)} = \color{tomato}{\textbf{1.050701}} \textit{ & } \textit{Negative Inputs Slope (}\alpha\textit{)} = \color{tomato}{\textbf{1.673263}} $

### SELU using PyTorch

```python
# Let's plot Pytorch's in-built SELU
selu = torch.nn.SELU()
y = selu(A)
plt.title("SELU Activation with PyTorch")
plt.plot(A.numpy(), y.numpy())
```

### Replicating SELU Activation

```python
def SELU(x, scale = 1.050701, alpha = 1.673263):
    return scale * torch.where(x > 0, x, alpha * (torch.exp(x) - 1))
```

```python
# Plot our SELU Activation
y_selu = SELU(A)
plt.title("Replicating SELU Activation")
plt.plot(A.numpy(), y_selu.numpy())
```
