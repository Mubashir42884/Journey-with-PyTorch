# **PYTORCH CV - MODEL BUILDING WORKFLOW**

### Computer Vision Model Workflow Overview

1. **Prepare Data and Turn into Tensors** ➜ Using `torchvision` utilities, like - `torchvision.datasets`, `torchvision.transforms`, `torchvision.ToTensor`, and `torch.utils.data.DataLoader`.


2. **Build a Model or Pick a pretrained Model** ➜ Using `torch` modules, like - `torch.nn`, `torch.nn.Module`, `torchvision.models`.
    * Pick a **loss function and optimizer** ➜ `torch.nn.Module` & `torch.optim`
    * Build a **training loop** 


3. **Making Predictions and Evaluation**  ➜ `model.eval` & `torchmetrics`


4. **Improving Baseline through Experimentations** ➜ more epoch, more linear layers, tuning learning rate, changing optimizer, etc...


5. **Save and Load the Model**  ➜ Using `torch.save` & `torch.load`

### 2. Build A Model

A baseline model is a simple model we will try and improve upon with subsequent models/experiments. Start with simple one and add complexity one by one.


#### Prelims of CV models -Flatten Layer:
* It flattens a contigious range of dims into a tensor. In a Convolutional Neural Network (CNN), a flatten layer **transforms multi-dimensional output from convolutional and pooling layers into a one-dimensional vector**, preparing it for input to fully connected layers.
From $$[channels, height, width] ➜ [channels, height * width]$$ $$[1, 28, 32] ➜ [1, 28 * 28] = [1, 784]$$

```python
# Create a flatten layer
flatten_model = nn.Flatten()

# Get a single sample
x = train_features_batch[0]
print(f"Sample Shape Before Flatten:\t{x.shape} ➜ [color_channels, height, width]")

# Flatten the sample
output = flatten_model(x)
print(f"Sample Shape After Flatten:\t{output.shape} \t➜ [color_channels, height*width]")
```

#### 2.1 Defining the Model

```python
from torch import nn
class FashionMNISTModel0(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
        super().__init__()
        self.layer_stack = nn.Sequential(
            # Add a flatten layer to flatten input image
            nn.Flatten(),
            
            # Add 1st linear layer, in➜input shape(img shape) and out➜ hidden units
            nn.Linear(in_features=input_shape,
                      out_features=hidden_units),
            
            # Add 2nd linear layer, in➜hidden units and out➜ output shape(class length)
            nn.Linear(in_features=hidden_units,
                      out_features=output_shape)
        )
        
    def forward(self, x: torch.Tensor):
        return self.layer_stack(x)
```

```python
torch.manual_seed(42)

# Setup model with input params
model_0 = FashionMNISTModel0(
    input_shape = 28*28,
    hidden_units= 10,
    output_shape= len(class_names)
).to('cpu')

# Create an instance of the model
model_0
```

##### 2.2.1 Let's beef up our model by creating a function to time our experiment
To experiment on our ML/DL/CV model, two major things we'll often want to track are:
1. Model's Performance (Loss, Accuracy, etc. evaluating values)
2. How fast it runs (The time to train the model to do the calculations)

```python
from timeit import default_timer as timer

## Create a function to time experiments
def print_train_time(start: float,
                    end: float,
                    device: torch.device = None):
    '''Prints difference between start and end time.'''
    total_time = end - start
    print(f"Train time on {device}: {total_time:.3f} seconds")
    return total_time
```

#### 2.3 Create Training Loop and Train Model on Batches
1. Loop through epochs
2. Loop through batches, perform training steps, calculate the train loss *per batch*
3. Loop through Tetsting batches, perform testing steps, calculate the test loss *per batch*
4. Print out what's happening
5. Time the model

```python
# Import tqdm for progress bar
from tqdm import tqdm

# set the seed and start the timer
torch.manual_seed(42)
train_time_start_on_cpu = timer()

# Set the number of epochs
epochs = 3

# Create training loop and test loop
for epoch in tqdm(range(epochs)):
    print(f"Epoch: {epoch}\n------")
    
    ## TRAINING
    train_loss = 0
    # Add a loop to loop through training batches
    for batch, (X, y) in enumerate(train_dataloader):
        model_0.train()
        '''1. Forward Pass'''
        y_pred = model_0(X)
        
        '''2. Calculate loss (per batch)'''
        loss = loss_fn(y_pred, y)
        train_loss += loss
        
        '''3. Optimizer zero grad'''
        optimizer.zero_grad()
        
        '''4. Loss Backward'''
        loss.backward()
        
        '''5. Optimizer Step'''
        optimizer.step()
        
        # Print out what's happening
        if batch % 400 == 0:
            print(f"Looked at {batch * len(X)} / {len(train_dataloader.dataset)} samples.")

    # Divide total train loss by length of train dataloader
    train_loss /= len(train_dataloader)
    
    ## TESTING
    test_loss, test_acc = 0, 0
    model_0.eval()
    with torch.inference_mode():
        for X_test, y_test in test_dataloader:
            '''1. Forward Pass'''
            test_pred = model_0(X_test)
            
            '''2.0 Calculate Loss'''
            test_loss += loss_fn(test_pred, y_test)
            
            '''2.1 Calculate Accuracy'''
            # using argmax to find out the maximum logits to compare labels to labels instead of logits
            test_acc += accuracy_fn(y_true=y_test, y_pred=test_pred.argmax(dim=1))
            
        # Calculate the test loss avg per batch
        test_loss /= len(test_dataloader)
        
        # Calculate the test acc avg per batch
        test_acc /=  len(test_dataloader)
        
    # Print out what's happening
    print(f"\nTrain Loss: {train_loss:.4f} | Test Loss: {test_loss:.4f} | Test Acc: {test_acc:.4f}")
    
# Calculate training time
train_time_end_on_cpu = timer()
total_train_time_model_0 = print_train_time(start=train_time_start_on_cpu, 
                                            end=train_time_end_on_cpu,
                                            device=str(next(model_0.parameters()).device))
```

### 3. Making Predictions and Evaluation

```python
torch.manual_seed(42)
def eval_model(model: torch.nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               accuracy_fn,
               device=None):
    """Returns a dict containing the results of model predicting on the DataLoader"""
    loss, acc = 0,0
    model.eval()
    with torch.inference_mode():
        for X, y in tqdm(data_loader):
            # Send data to the target device
            X, y = X.to(device), y.to(device)
            # Make predictions
            y_pred = model(X)
            
            # Accumulate loss and acc values per batch
            loss += loss_fn(y_pred, y)
            acc += accuracy_fn(y_true=y, 
                               y_pred=y_pred.argmax(dim=1))
            
        # Scale loss and acc to find the avg loss/acc per batch
        loss /= len(data_loader)
        acc /= len(data_loader)
        
    return {"model_name": model.__class__.__name__, # only works when model was created with a class 
            "model_loss": loss.item(),
            "model_acc": acc}
```

```python
# Calculate the model results on test dataset
model_0_results = eval_model(model=model_0,
                            data_loader=test_dataloader,
                            loss_fn=loss_fn,
                            accuracy_fn=accuracy_fn,
                            device='cpu')

model_0_results
```

#### 4.2 Create a Model with Non-linear and Linear Layers

```python
from torch import nn
class FashionMNISTModel1(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
        super().__init__()
        self.layer_stack = nn.Sequential(
            # Add a flatten layer
            nn.Flatten(),
            
            # Add 1st linear layer
            nn.Linear(in_features=input_shape,
                      out_features=hidden_units),
            
            # Add non-linear layer
            nn.ReLU(),
            
            # Add 2nd linear layer
            nn.Linear(in_features=hidden_units,
                      out_features=output_shape),
            
            # Add non-linear layer
            nn.ReLU(),
        )
        
    def forward(self, x: torch.Tensor):
        return self.layer_stack(x)
```

```python
# Create an instance of Model 1
torch.manual_seed(42)
model_1 = FashionMNISTModel1(
    input_shape=28*28,
    hidden_units=10,
    output_shape=len(class_names)
).to(device)

# Create an instance of the model
model_1
```

###### Bonus: Turning the Training loop and Testing loop into a Function

```python
# Create a function for training and evaluating

def training_step(model: torch.nn.Module,
                  data_loader: torch.utils.data.DataLoader,
                  loss_fn: torch.nn.Module,
                  optimizer: torch.optim.Optimizer,
                  accuracy_fn,
                  device: torch.device = device):
    """
    Executes a single training step for a given model.

    Params:
        model:       The model to be trained.
        data_loader: DataLoader providing the training data.
        loss_fn:     Loss function to calculate the training loss.
        optimizer:   Optimizer to update model parameters.
        accuracy_fn: Function to calculate accuracy during training.
        device:      Device to perform computations on 'cpu' or 'cuda' (Default current device)

    Returns:
        float: Average training loss over all batches in the DataLoader.
        float: Average training accuracy over all batches in the DataLoader.
    """
    
    train_acc, train_loss = 0, 0
    
    '''Put model on the train mode'''
    model.train()
    
    for epoch in tqdm(range(epochs)):
        print(f"Epoch: {epoch}\n------")

    # Add a loop to loop through training batches
    for batch, (X, y) in enumerate(data_loader):
        '''Put data on the target device'''
        X, y = X.to(device), y.to(device)
        
        '''1. Forward Pass'''
        y_pred = model(X)
        
        '''2. Calculate loss and accuracy (per batch)'''
        loss = loss_fn(y_pred, y)
        train_loss += loss
        train_acc += accuracy_fn(y_true=y,
                                 y_pred=y_pred.argmax(dim=1))
        
        '''3. Optimizer zero grad'''
        optimizer.zero_grad()
        
        '''4. Loss Backward'''
        loss.backward()
        
        '''5. Optimizer Step'''
        optimizer.step()

    # Divide total train loss and accuracy by length of train dataloader
    train_loss /= len(data_loader)
    train_acc /= len(data_loader)
    print(f"| Train Loss: {train_loss:.4f}  |  Train Acc: {train_acc:.4f}% |")
```

```python
# Create a function for training and evaluating

def test_step(model: torch.nn.Module, 
              data_loader: torch.utils.data.DataLoader, 
              loss_fn: torch.nn.Module, 
              accuracy_fn,
              device: torch.device = device):
    """
    Executes a single test step for a given model.

    Params:
        model:       The model to be trained.
        data_loader: DataLoader providing the test data.
        loss_fn:     Loss function to calculate the test loss.
        accuracy_fn: Function to calculate accuracy during test.
        device:      Device to perform computations on 'cpu' or 'cuda' (Default current device)

    Returns:
        float: Average test loss for the step.
        float: Average test accuracy for the step.
    """
    
    test_acc, test_loss = 0, 0
    
    '''Put model on the eval mode'''
    model.eval()
    
    # Turn on inference mode context manager
    with torch.inference_mode():
        for X, y in data_loader:
            # Send data to the target device
            X, y = X.to(device), y.to(device)
            
            '''1. Forward Pass'''
            test_pred = model(X)
            
            '''2. Calculate the loss'''
            test_loss += loss_fn(test_pred, y)
            test_acc += accuracy_fn(y_true=y,
                                    y_pred=test_pred.argmax(dim=1))
            
        # Adjust metrics and print results
        test_loss /= len(data_loader)
        test_acc /= len(data_loader)
        print(f"| Test Loss: {test_loss:.4f}  |  Test Acc: {test_acc:.4f}% |\n")
```

```python
torch.manual_seed(42)

# Measure time
train_time_start_on_gpu = timer()
epochs = 3

# Create an optimization and evaluation loop using train_step() and test_step()
for epoch in tqdm(range(epochs)):
    print(f"\tEpoch: {epoch} \n------------")
    ## TRAIN STEPS
    training_step(model=model_1,
                  data_loader=train_dataloader,
                  loss_fn=loss_fn,
                  optimizer=optimizer,
                  accuracy_fn=accuracy_fn,
                  device=device)
    
    ## TEST STEPS
    test_step(model=model_1,
              data_loader=test_dataloader,
              loss_fn=loss_fn,
              accuracy_fn=accuracy_fn,
              device=device)
    
# Calculate training time
train_time_end_on_gpu = timer()
total_train_time_model_1 = print_train_time(start=train_time_start_on_gpu, 
                                            end=train_time_end_on_gpu,
                                            device=device)
```
