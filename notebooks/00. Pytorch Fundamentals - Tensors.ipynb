# What is a Tensor?

A tensor is a multi-dimensional numerical array that generalizes scalars, vectors, and matrices. Key characteristics:
- Fundamental data structure in ML/DL
- Represents data across 0 to n dimensions
- Container for numerical computations
- Supports complex mathematical operations

# Why Tensor is Important in ML/DL?

Tensors are critical because they:
1. Enable efficient parallel processing
2. Represent complex multi-dimensional data naturally
3. Support GPU acceleration
4. Facilitate gradient computation
5. Provide a standardized data format across frameworks

#### SCALER

```python
scaler = torch.tensor(5)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype} \t# by default int64")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")

scaler = torch.tensor(5.3)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype}")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")

scaler = torch.tensor(True)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype}")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")

scaler = torch.tensor(1+3j)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype}")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")
```

```python
scaler = torch.tensor(5, dtype=torch.float32)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype} \t# changing dtype explicitly to float")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()} \t# only works for single element tensors(scalers) \n\n")

scaler = torch.tensor(5, dtype=torch.complex64)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype} \t# changing dtype explicitly to complex")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")

scaler = torch.tensor(5.3, dtype=torch.int32)
print(scaler)
print(f"Scaler datatype\t: {scaler.dtype} \t# changing dtype explicitly to complex")
print(f"Scaler shape\t: {scaler.shape}") 
print(f"Scaler dimension: {scaler.ndim}")
print(f"Scaler item\t: {scaler.item()}\n\n")
```

#### VECTOR

```python
vector = torch.tensor([5,9,12])
print(vector)
print(f"\nVector datatype\t: {vector.dtype}")
print(f"Vector shape\t: {vector.shape}") 
print(f"Vector dimension: {vector.ndim}")
print(f"Vector item 1\t: {vector[0]}")
print(f"Vector item 2\t: {vector[1]}")
print(f"Vector item 3\t: {vector[2]}")
```

#### MATRIX

```python
mat = torch.tensor([[4, -1, 3],
                    [5, 12, 9],
                    [11, -5, 7]])
print(mat)
print(f"\nMatrix datatype\t: {mat.dtype}")
print(f"Matrix shape\t: {mat.shape}\t #it's a 3x3 matrix") 
print(f"Matrix dimension: {mat.ndim}\t\t\t #it's has two dims: row and column")
print(f"Matrix Row 1\t: {mat[0]}")
print(f"Matrix Row 2\t: {mat[1]}")
print(f"Matrix Row 3\t: {mat[2]}\n")

print(f"Matrix Item (0,2)\t: {mat[0][2]}")
print(f"Matrix Item (1,2)\t: {mat[1][2]}")
print(f"Matrix Item (2,1)\t: {mat[2][1]}")
```

#### TENSORS

```python
tensors = torch.tensor([[[2, 3, 5],
                         [4, 1, 9],
                         [5, 3, 0],
                         [7, 5, 6]],
                         [[5, 2, 4],
                         [9, 8, 7],
                         [1, 0, 3],
                         [8, 4, 6]]])
print(tensors)
print(f"\nTensor datatype\t: {tensors.dtype}")
print(f"Tensor rank \t: {tensors.ndim}")  # Total number of dimensions


dim0, dim1, dim2 = tensors.shape

print(f"dim0 (number of matrices) \t: {dim0}")
print(f"dim1 (rows per matrix)    \t: {dim1}")
print(f"dim2 (columns per row)    \t: {dim2}")

print(f"\nTensor shape\t: {tensors.shape}\t # [dim0, dim1, dim2]")
print(f"Tensor dimension: {tensors.ndim}\t\t\t # 3D Tensor\n")

print(f"Tensor 2nd Matrix 1st Row\t: {tensors[1][0]}")
print(f"Tensor 1st Matrix 2nd Row\t: {tensors[0][1]}")
print(f"Tensor 2nd Matrix 3rd Row\t: {tensors[1][2]}")
print(f"Tensor 1st Matrix 4th Row\t: {tensors[0][3]}\n")

print(f"Tensor Item (0,3,0)\t: {tensors[0][3][0]} \t# 1st Matrix, 4th row, 1st item")
print(f"Tensor Item (0,1,2)\t: {tensors[0][1][2]} \t# 1st Matrix, 2nd row, 3rd item")
print(f"Tensor Item (1,2,2)\t: {tensors[1][2][2]} \t# 2nd Matrix, 3rd row, 3rd item")
print(f"Tensor Item (1,0,2)\t: {tensors[1][0][1]} \t# 2nd Matrix, 1st row, 2nd item")
```

### Random Tensors

Random Tensors are important in Neural Networks because it start learning from a random numbers and then adjust those in learnt representation.

```python
# Creating Random Tensors

rd_tensors = torch.rand(2,3)
print(rd_tensors)
print(f"Dimension: {rd_tensors.ndim}")
print(f"Shape\t : {rd_tensors.shape}\n")

rd_tensors = torch.rand(1,3,4)
print(rd_tensors)
print(f"Dimension: {rd_tensors.ndim}")
print(f"Shape\t : {rd_tensors.shape}")
```

### How a tensor of image should look like?

An image has three attributes: color channels (R,G,B), height, and width.</br>
So, an image with 224x224 size should have = `(3, 224, 224)` shaped tensors.

```python
image = torch.rand(size=(3,224,224)) # Channel-first format in PyTorch
print(image)
print(f"\nImage Shape\t: {image.shape}")
print(f"Image Dimension : {image.ndim}")
```

```python
import matplotlib.pyplot as plt

# Convert to channel-last format for Matplotlib
image = image.permute(1, 2, 0).numpy()

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

### Range of Tensors and Tensor-like

```python
# Use torch.arange()

rng = torch.arange(0,10)
print(rng)
print(rng.shape)

step_range = torch.arange(start=5, end=1000, step=150)
print(step_range)
print(step_range.shape)

print("\nSimilar shaped Tensor like `step_range` tensor but with zeros:")
tensorlike = torch.zeros_like(input=step_range)
print(tensorlike)
print(tensorlike.shape)
```

### Errors and Issues in working with Tensor
1. Tensors are not in right datatype `torch.dtype`
2. Tensors are not in right shape `torch.shape`
3. Tensors are not in right device `torch.device`

```python
data = torch.rand(size=(4,3),
                  dtype=None,
                  device=None,
                  requires_grad=False)

print(data)
print(f"\nDatatype of Tensor:\t{data.dtype}")
print(f"Shape of Tensor:\t{data.shape}")
print(f"Device of Tensor:\t{data.device}")
```
